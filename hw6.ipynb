```python
import numpy as np
import matplotlib.pyplot as plt

# Part a

n = 5000

# Generate 5000 uniform samples of U1 and U2
U1 = np.random.rand(n)
U2 = np.random.rand(n)

# Apply the Box-Muller transformation to get X1 and X2
X1 = np.sqrt(-2 * np.log(U1)) * np.cos(2 * np.pi * U2)
X2 = np.sqrt(-2 * np.log(U1)) * np.sin(2 * np.pi * U2)


```

```python
# Part b
# Plot the histogram of X1 and X2

from scipy.stats import norm

# Plot the normalized histogram of X1. The parameter 'density=True' scales the histogram so that its area is 1.
plt.hist(X1, bins=40, density=True, alpha=0.6, color='skyblue', edgecolor='black', label='X1')

# Generate x values spanning the range of our data for plotting the PDF.
xmin, xmax = plt.xlim()  # Get the current x limits of the plot
x = np.linspace(xmin, xmax, 100)  # Generate 100 evenly spaced points between xmin and xmax

# Plot the PDF of the standard normal distribution
p = norm.pdf(x, 0, 1)  # Compute the PDF of the standard normal distribution at the points x
plt.plot(x, p, 'r', linewidth=2, label='Standard Normal PDF')  # Plot the PDF

plt.title('Histogram of X1 and Standard Normal PDF')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.show()

plt.figure()

# Plot the normalized histogram of X1. The parameter 'density=True' scales the histogram so that its area is 1.
plt.hist(X2, bins=40, density=True, alpha=0.6, color='skyblue', edgecolor='black', label='X2')

# Generate x values spanning the range of our data for plotting the PDF.
xmin, xmax = plt.xlim()  # Get the current x limits of the plot
x = np.linspace(xmin, xmax, 100)  # Generate 100 evenly spaced points between xmin and xmax

# Plot the PDF of the standard normal distribution
p = norm.pdf(x, 0, 1)  # Compute the PDF of the standard normal distribution at the points x
plt.plot(x, p, 'r', linewidth=2, label='Standard Normal PDF')  # Plot the PDF

plt.title('Histogram of X2 and Standard Normal PDF')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.show()

```

```python
# Part c 
# Plot the samples of X1 vs X2 in a scatter plot

plt.figure()
plt.scatter(X1, X2, alpha=0.8, color='purple', s = 10)
plt.title('Scatter plot of X1 vs X2')
plt.xlabel('X1')
plt.ylabel('X2')
plt.grid(True)
plt.show()
```
