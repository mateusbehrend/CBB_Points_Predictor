```python
import pandas as pd 
matches = pd.read_csv("miles-byrd_game_log.csv")
matches
```

```python
ken_pom = pd.read_csv("kenpom.csv")
ken_pom.head()
```

```python
from fuzzywuzzy import process
# Function to find best matching state for each team_name
def match_state(team_name, states):
    match, score = process.extractOne(team_name, states)
    # Set a threshold for matching (e.g., 80 out of 100)
    return match if score >= 80 else None

# Apply the function to create a matching column
ken_pom['state'] = ken_pom['Team'].apply(lambda x: match_state(x, matches['Full_Team_Name'].tolist()))

ken_pom['state'].value_counts()
```

```python
# For each opponent team get the matching ken_pom row
ken_pom_opponents = ken_pom[ken_pom['state'].notnull()]
ken_pom_opponents
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=100, min_samples_split=5, random_state=1)
```

```python
merged_df = pd.merge(matches, ken_pom_opponents, left_on="Full_Team_Name", right_on="state", how="left")
merged_df
```

```python
merged_df['DATE'] = pd.to_datetime(merged_df['DATE'], format='%a %m/%d')

# Then apply your lambda function to update the year
merged_df['DATE'] = merged_df['DATE'].apply(
    lambda dt: dt.replace(year=2024) if dt.month in [11, 12] else dt.replace(year=2025)
)
```

```python

train = merged_df[merged_df["DATE"] < "2025-02-01"]
train
```

```python
test = merged_df[merged_df["DATE"] >= "2025-02-01"]
test
```

```python
column_names = list(merged_df.columns)
print(merged_df.columns.tolist())
```

```python
# predictors = ["MIN", "AST", "REB", "TO", "Home", "Away", "NetRtg", "ORtg", "DRtg", "AdjT"]
predictors = ["Home", "Away", "NetRtg", "ORtg", "DRtg", "AdjT", "R_Tempo", "Rk"]
```

```python
rf.fit(train[predictors], train["PTS"])
```

```python
preds = rf.predict(test[predictors])
```
